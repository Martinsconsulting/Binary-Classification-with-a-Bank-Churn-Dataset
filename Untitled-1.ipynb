```python
def viz(x,y):
    df_viz = df.groupby(x)[y].value_counts()
    df_viz = df_viz.unstack()
    df_viz.columns= ['Stayed', 'Exited']
    df_viz.plot(kind='bar')
    plt.xlabel('Country')
    plt.ylabel('Counts')
    plt.title(f'Exited and non Exited accounts per {x}')
    plt.legend()
    plt.show()
    df_viz.reset_index(inplace=True)
    df_viz['Total'] = df_viz['Exited'] + df_viz['Stayed']
    df_viz[f'% of Stayed'] = df_viz['Stayed'] / df_viz['Total']*100
    df_viz[f'% of Exited'] = df_viz['Exited'] / df_viz['Total']*100
    return df_viz
```

```python
def viz_hist(feature):
    column = df[feature]
    exited = df['Exited']

    plt.hist([column[exited==0], column[exited==1]], bins=10, color=['skyblue', 'salmon'], label=['Retained', 'Closed'])
    plt.xlabel(feature)
    plt.ylabel('Counts')
    plt.title(f'Histogram of Credit Scores by {feature}')
    plt.legend()
    plt.show()
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
encoded = encoder.fit_transform(df[['Geography', 'Gender']])
encoded_df = pd.DataFrame(encoded.toarray(), columns=encoder.get_feature_names_out(['Geography', 'Gender']))

df_encoded = pd.concat([df, encoded_df], axis=1)
df_encoded.drop(columns=['Geography', 'Gender'], axis=1, inplace=True)
df_encoded
```

```python
features = df_encoded.drop(columns=['Exited', 'id', 'Balance'])
targets = df_encoded['Exited']
```

```python
features
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_clf = RandomForestClassifier()
rf_clf.fit(features, targets)
importances = rf_clf.feature_importances_

importances_df = pd.DataFrame({'Feature': features.columns, 'Importance': importances})
importances_df = importances_df.sort_values(by='Importance', ascending=True)

plt.barh(importances_df['Feature'], importances_df['Importance'])
plt.xlabel('Feature')
plt.ylabel('Importance')
plt.show()
```

```python
importances_df['Importance'] = importances_df['Importance']*100
importances_df
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import roc_auc_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, targets, random_state=42)
```

```python
data ={'Logistic':LogisticRegression,
       'KNeighbors': KNeighborsClassifier,
       'Tree' : DecisionTreeClassifier,
       'Random Forest': RandomForestClassifier,
       'Gradient Boosting': GradientBoostingClassifier,
       'SVC': SVC,
       'MLP': MLPClassifier,
       }

model_scores = {}
for name, model in data.items():
    clf = model()
    clf.fit(X_train, y_train)
    if name == 'SVC':
        roc_auc = clf.decision_function(X_test)
    else: 
        pred = clf.predict_proba(X_test)[:,1]
    roc_auc = roc_auc_score(y_test, pred)
    model_scores[name] = {'roc_auc_score': roc_auc}

data = [{'Model': model_name, **scores} for model_name, scores in model_scores.items()]

df_model = pd.DataFrame(data)
df_model.sort_values(by='roc_auc_score', ascending=False)
```

```python
param_grid = {
    'n_estimators': [50,100,150,200],
    'learning_rate': [0.01,0.1,1,5],
    'max_features': [1,2,3,4,5]
}

clf = GradientBoostingClassifier(random_state=42)
grid = GridSearchCV(estimator=clf, param_grid=param_grid, cv=3, scoring='roc_auc')
grid.fit(X_train, y_train)
print('Best Parameters:', grid.best_params_)
```

```python
clf = grid.best_estimator_
y_proba = clf.predict_proba(X_test)[:,1]
roc_auc = roc_auc(y_test,y_proba)
print('roc_auc_score on test set ',roc_auc)
```
